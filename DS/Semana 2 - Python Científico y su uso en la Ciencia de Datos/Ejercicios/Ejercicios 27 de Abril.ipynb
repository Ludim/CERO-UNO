```python
#Ejercicio 10. Del siguiente areglo [1,2,0,0,4,] encuentra los índices 
#donde los elementos sean diferente a cero.

test=[1,2,0,0,4,]
indices=[]
for pareja in list(enumerate(test)):
    if pareja[1]!=0:
        indices.append(pareja[0])
print(indices)
```

```python
#Crea una matriz de 5x5 
#con valores 1, 2, 3, 4 justo después de la diagonal.

claro=np.eye(5)
print(claro)
print()
claro2=claro
i=1
while i<=4:
        claro2[i][i-1]=i
        i+=1
print(claro2)
```

```python
#Ejercicio 15. Normaliza una matriz aleatoria de 5x5

matriz=np.array([np.random.randint(2,6) for i in range(25)]).reshape(5,5)
normalizacion=matriz/np.linalg.det(matriz)
print(matriz)
print()
print(normalizacion)
```
