### 1.   Crea un vector con 100 números aleatorios uniformemente distribuidos en el intervalo comprendido entre el -1 y 1

### 3. Crea una matriz que tenga 10 filas y 10 columnas con 100 números aleatorios entre -1 y 1.

```
# No hay matrices, pero existen listas de listas
matriz_100 = []

for i in range(10):
  matriz_100.append(random_vector[10*i:10*i+10])

print(matriz_100)
```

### 4. Imprime la tercer columna de la matriz creada en el ejercicio 3

```
#Si usamos un solo índice obtenemos solo la fila 
print('La primera fila de la matriz es:')
print(matriz_100[0])
print()


# La tercera columna representa el tercer elemento de cada fila 
ter_col=[matriz_100[i][2] for i in range(10)]
print('La tercera columna es:')
print(ter_col)
```

### 7.   Crea un vector del 1 al 5 y eleva a la potencia 3 cada elemento.

### 9.   Calcular el producto punto de los siguientes dos vectores:  [1,2,3] y  [3,2,1]

```
# Definimos los vectores a usar 
a = [1,2,3]
b = [3,2,1]

# Usando tuplas y un ciclo
punto = 0
for e, f in zip(a,b):
  punto += e*f
print('El producto punto es '+str(punto))
print()

# Usando List comprehensions 
punto2=sum([a[i]*b[i] for i in range(len(a))])
print('El producto punto es '+str(punto2))
```

# 2. Numerical Python (NumPy)


*Numpy* o *Numertical Python* es un paquete fundamental sobre el cual estan construidos muchos de los paquetes de ciencia de datos más utilizados. Su elemento fundamental son los *arrays* (arreglos) los cuales toman características operacionales similares a las matrices y vectores de álgebra lineal. 

Algunas operaciones que se pueden realizar: sumar, restar, multiplicar, producto punto, inversa de una matriz, sistemas lineales, etc.

### 2.1 ¿Cómo usamos Numpy?

```
import numpy as np
```

### 2.2 ¿Qué es un vector?

Un vector es... 
- Conjunto de elementos del mismo tipo
- Matriz unidimensional
- Mutable

La manera más fácil de crear un arreglo es convertir a partir de una lista predeterminada usando el comando `array`

```
#Arreglo unidimensional 
arreglo=np.array([3, 33, 333])
print(arreglo)
print()

# Método shape 
print(arreglo.shape) 
print()

# Método size
print(arreglo.size)
```

Ahora definamos un arreglo bidimensional, algo parecido a una matriz: ahora tenemos renglones y columnas. 

De igual manera podemos definirlo de igual manera con el comando `array` y una serie de listas que seran interpretadas como las filas de nuestros arreglos

```
#Arreglo bidimensional
arreglo2D=np.array([[11,12,13],[21,22,23]])
print(arreglo2D)
print()

# Método shape 
print(arreglo2D.shape) 
print()

# Método size
print(arreglo2D.size)
print()

print(arreglo2D.ndim)
```

Ahora exploremos algunos tipos de arreglos preeterminados: busquemos como generar de manera rápida mendiante funciones de NumPy los siguientes arreglos: 

* únicamente contiene ceros

* únicamente contiene unos 

* únicamente contiene valores iguales a una constante

* que contenga únicamente unos en la diagonal principal y ceros en las demás posiciones

* que sea cuadrado y que contenga únicamente unos en la diagonal principal y ceros en las demás posiciones

* contenga números aleatorios uniformemente distribuidos

* contenga los enteros en un cierto intervalo específico

* contenga N números uniformemente espaciados en un cierto intervalo. 






##### Arreglos preestablecidos

```
# Arreglo que solo contiene ceros
cero=np.zeros((3,2))
print(cero)
```

```
# Arreglo que solo contiene ceros
uno=np.ones((2,3))
print(uno)
```

```
# Arreglo de valores en un intervalo que estan igualmente espaciados
lin=np.linspace(0,4,10)
print(lin)
```

### 2.3 Operaciones


#### 2.3.1 Aritmética 

A diferencia de las listas, donde tenemos que hacer algun proceso para operar cada elemento, en los arreglos las operaciones se realizan para cada elemento.

```
# Creamos unos arreglo para unas pruebas
A=np.arange(10)
B=np.arange(2,12)

print('Arreglo original A=',A)
print('Arreglo original B=',B)
print()

# Suma de un elemento 
print('Suma de una costante ',A+1)
print()

#Suma de arreglos 
print('Suma de arreglos A+B=',A+B)
print()

#Resta
print('Diferencia de arreglos',A-B)
print()


#Multiplicación
print('Multiplicación A*B=',A*B)
print()

#Division
print('Divisio A/B=',A/B)
print()

#Potencias  
print('Elevar el arreglo a la segunda potencia A**2=',A**2)
print()

print('Elevar un numero a la potencia indicada por el arreglo 2**A ',2**A)
print()

print('Elevar un arrelgo a la potencia indicada por el otro arreglo B**A ',B**A)
```

Numpy posee las funciones estadísiticas básicas: 
* Media

Para mayor información: https://docs.scipy.org/doc/numpy-1.14.5/reference/routines.statistics.html

```
# Creamos un arreglo unidimensional de 0 al 19
arr = np.arange(20)
print('Original')
print(arr)
print()

# Le cambiamos la forma para que sea un arreglo de 4 x 5 
print('Reacomodo')
ARR=arr.reshape(4,5)
print(ARR)
```

```
# Transposición: cambio de columnas por filas y viceversa
print(ARR.T)
```

```
# Aplanamos el arreglo de dos dimensiones para que sea de nuevo de uno
print(np.ravel(ARR))
```

```
#Creamos dos arreglos de 2 dimensiones que solo contengan enteros aleatoriamente 
K = np.random.randint(low=2,high=50,size=(2,2))
print('K=')
print(K)
print()
print('M=')
M = np.random.randint(low=2,high=50,size=(2,2))
print(M)
```

```
# Concatenación vertical

#Forma 1
VS=np.vstack((K,M))

#Forma 2 
VS2=np.concatenate([K, M], axis = 0)


print('Concatenado vertical')
print(VS)
print()
print(VS2)
```

```
# Concatenación horizontal

#Forma 1
VH=np.hstack((K,M))

#Forma 2 
VH2=np.concatenate([K, M.T], axis = 1)

print('Concatenado horizontal')
print(VH)
print()
print(VH2)
```

#### 2.3.6 Broadcasting 

Una de las ventajas que nos ofrece numpy es poder operar elementos de arreglos de diferentes tamaños y esto se llama broadcasting. 

Para mayor información: https://docs.scipy.org/doc/numpy-1.10.1/user/basics.broadcasting.html

```
A = np.zeros((4,3))
print(A)
```

```
#Suma en cada fila los elementos del vector 

print('Vector renglon VR',np.array([2,4,1]))
print()
print('Operacion A+VR ')
print(A+np.array([2,3,1]))
print()


#Ahora suma las columnas
print('Vector columna VC')
print(np.array([[2],[3],[1],[0]]))
print()
print('Operacion A+VC ')
print(A+np.array([[2],[3],[1],[0]]))
```
