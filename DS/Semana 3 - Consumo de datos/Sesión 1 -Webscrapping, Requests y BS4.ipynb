Cuando obtenemos información de una página de internet esta no se encuentra en el formato tradicional en el cual estamos acostumbrados (excel, tablas, sql, etc). El formato de la información puede parecer confuso pero la verdad es que al obtener la información de la página recuperamos el código que se encuentra detrás de esta que puede ser html, xml o javascript por lo cual es importante saber como se comporta esta información para poder navegarla de forma fácil y poder aprovechar al máximo esta información. Para ello podemos categorizar las bases de información en 2 formas distintas, bases de datos relacionales y las bases de datos no relacionales.

## Base de datos relacionales

"Una base de datos relacional es aquella que está organizada en renglones y columnas dentro de una tabla." Es decir, es una tabla cuyas columnas representan atributos de cierta medición (Edad, precio, genero, especie, tipo de soldadura) y en cada renglón tenemos una tupla correspondiente a una observación. Las bases de datos relacionales siguen el formato de SQL y se pueden tratar con mucha facilidad pero al precio de su velocidad. Una mejor forma de ver esto es como una spreadsheet o una tabla de excel en donde los renglones corresponden a observaciones mientras que las columnas son características de esa observación. Claro que puede haber otras tablas con variables similares pero en esencia todo se conserva en un mismo nivel y no cuentas con cosas anidadas.

## Bases de datos no relacionales

Todas las bases de datos que sigan una jerarquía son definidas como bases de datos no relacionales; un ejemplo serían los diccionarios anidados en donde tienes que recorrer distintos niveles para tener acceso a diferentes tipos de información. Asimismo podemos ver que el código html es no relacional ya que hay diferentes niveles y atributos en donde se accede a diferentes tipos de información como puede ser el usuario o el post o la hora en caso de un foro. En lo particular, la "Big Data" se basa en bases de datos no relacionales debido a la gran cantidad de fuentes de información que tienen (carros, sensores, refrigeradores, streaming de páginas). Estas son las bases de datos más comunes pero a pesar de su orden distinto y falta de relación directa, estas se pueden llegar a convertir en bases de datos relacionales si uno conoce las herramientas y las transformaciones necesarias para lograrlo. De hecho; este es nuestro objetivo.

## Requests
Cuando queremos acceder a una página de internet nosotros tecleamos la dirección en el navegador; después, nuestro navegador hace una petición al servidor para recibir la información acerca de la página a la que queremos acceder. Un request es una petición para entrar a la página de nuestro interés; ahora que se hace un request a la página se debe de decidir el método que se va a realizar.

Lo primero que podemos hacer es obtener la información, esto se hace con el get. Es una solicitud para recuperar el código de la página de internet. Esté método se puede utilizar en python importando el paquete de requests; al correr el método dir sobre el objeto request podemos obtener una lista que nos dice la información y las etiquetas de la página que podemos acceder.


```python
import requests
page = requests.get('http://www.amazon.com')
dir(page)
```

Como podemos ver, recuperamos un objeto muy grande que puede tener mucha información pero esta no se encuentra tan a la vista como lo puede ser en un documento word. Si queremos acceder a esta información tenemos que utilizar la notación de Python de .objeto.

```python
print(page.url, page.request, page.encoding)
```

Aunque si vemos la página, podemos ver que no tenemos los permisos apropiados para scrapear amazon y nos regresa un mensaje automatizado para poder scrapearla con su permiso. Muchos sitios web tienen en pie este tipo de restricciones por lo que es importante investigar antes de tiempo cuales sitios soportan el scrapping.

```python
page.text
```

Otra opción es publicar información; ya sea crearla o actualizarla. Esto se hace por medio del método post (sigue siendo un request); con esta se manda una petición a un sitio web de práctica para agregar información al segmento de form. 

```python
r = requests.post('http://httpbin.org/post', data = {'Nombre':'Yo'})
r.content
```

Regresando al método get; nosotros podemos poner una payload que añade llaves y valores al url. Hay ciertas páginas que tienen un comportamiento predecible y esto permite poder scrapearlas de forma automatizada.

```python
payload = {'key1': 'value1', 'key2': ['value2', 'value3']}
r = requests.get('http://httpbin.org/get', params=payload)
print(r.url)
```

También podemos mandar solicitudes para borrar información o solamente probar la respuesta del servidor; estos son los métodos delete y head. Para scrapping no los usaremos mucho ya que en si no queremos modificar la información del documento sino queremos extraerla.

# BeautifulSoup4
Ahora; los nombres de los paquetes de Python muchas veces se basan en obras de la cultura popular. Esto surge de Alicia en el país de las maravillas debido a las personas que hacen el programa se reservan los derechos de nombrarlo. Ahora bien; esta libreria es muy similar a la de requests ya que podemos hacer los requests tradicionales; pero la diferencia está en que con este paquete ustedes pueden navegar la estructura de la pagina web con mucha mayor facilidad que utilizando Python vanilla. Esto nos permite extraer la información de la página web y utilizarla para nuestras propias cuestiones.

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = 'https://www.milenio.com'
URL = BeautifulSoup(requests.get(URL).text, "lxml")
dir(URL)
```

Este objeto no es un código html, es lo que se le conoce como una sopa (tiene muchos ingredientes pero los tenemos bien catalogados). Con estos comandos podemos reobtener parámetros interesantes como con el método de requests pero en este caso podemos explorar muchas cosas más.

```python
URL.title, URL.title.name, URL.title.string, URL.index
```

El método de find_all nos permite buscar el tag del renglón incluyendo el texto y las demás propiedades.

```python
URL.find_all('a')
```

Asimismo se pueden extraer los subenlaces que llevan a diferentes regiónes de la página.

```python
for a in URL.find_all('a', href=True):
    print ("Found the URL:",a.get_text().strip(), a['href'])
```

Si el renglón cuenta con una etiqueta y una clase, esta se puede especificar dentro del método junto con atributos como href. En el siguiente ejemplo accedemos a la etiqueta li y la clase father para extraer el primer elemento.

```python
lista=URL.find_all('li',class_='father')
print(lista[0])
```

De esta lista podemos extraer el texto con el método get_text(). Usamos el método .strip() para quitar los espacios y tabs adicionales que puede tener el formato del texto.

```python
# Copiamos los resultados
Categorias=[str(lista[i].get_text()).strip() for i in range(0,len(lista)-1)]
```

```python
print(Categorias)
```

Una página que si nos deja scrapear es Soriana; a pesar de que podemos scrapearla no hay que abusar ya que podemos terminar haciendo que gasten muchos recursos por lo que solo podemos usarla para propósitos educativos. Primero vamos a intentar extraer los productos de repostería y gourmet junto con sus precios correspondientes; investigando en la página vemos que los productos tienen la etiqueta <div class='product-item' con lo cual podemos acceder a la informacion de cada producto. Navegando los niveles dentro de esto encontramos cosas como el nombre del producto y el precio.

```python
URL2 = 'https://www.soriana.com/soriana/es/c/Hogar/Cocina/Reposteria-y-gourmet/c/36C'
URL2 = BeautifulSoup(requests.get(URL2).text, "lxml")
```

```python
produ=URL2.find_all('div',class_='product-item')
print(produ[0].find_all('div',class_='priceContainer'))
```

Usando iteradores; podemos extraer todos los datos que necesitamos y hacer una lista.

```python
N=[link.get_text().strip() for item in URL2.find_all('div',class_='product-item') for link in item.find_all('a')]
N=list(filter(lambda x: x!='',N))
print(N)
P=[link.get_text().strip() for item in URL2.find_all('div',class_='product-item') for link in item.find_all('span',class_='price')]
print(P)
```
