```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(np.random.normal(3,1,size=(1000, 4)), columns=list('ABCD'))
df.head()
```

```python
hist=df.hist(bins=20,layout=(2,2))
plt.axis('tight')
```

```python
hist=df.hist(bins=10,column=['A','B'],layout=(1,2))
plt.axis('tight')
```

```python
df['y']= (pd.Series(3*df['A'], index=df.index))
df.head()
```

```python
df.plot.scatter(x='A',y='B',c='g')
```

```python
df.plot.scatter(x='A',y='y')
```

```python
Matrix=pd.plotting.scatter_matrix(df, alpha=0.3)
```

```python
corr = df.corr()
print(corr)
```

```python
corr = df.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```
